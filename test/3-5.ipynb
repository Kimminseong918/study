```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

import os

if os.name == 'posix':

    plt.rc("font", family="AppleGothic")

else:
    plt.rc("font", family="Malgun Gothic")
    
plt.rc("axes", unicode_minus=False)

# df = pd.read_csv("boostcourse-ds-510/data/NHIS_OPEN_GJ_2017.CSV.zip", encoding="cp949")
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip",
                 encoding="cp949")
```

```python
df.head()
```

```python
df_sight= df[df['시력(좌)'] != 9.9]
df_sight= df_sight[df_sight['시력(우)'] != 9.9]
df_sight[["시력(좌)", "시력(우)"]].describe() # 실명 데이터 제거 완료
```

```python
age_code = {1: '0~4세',
 2: '5~9세',
 3: '10~14세',
 4: '15~19세',
 5: '20~24세',
 6: '25~29세',
 7: '30~34세',
 8: '35~39세',
 9: '40~44세',
 10: '45~49세',
 11: '50~54세',
 12: '55~59세',
 13: '60~64세',
 14: '65~69세',
 15: '70~74세',
 16: '75~79세',
 17: '80~84세',
 18: '85세+'}
```

```python
df['연령대'] = df['연령대코드(5세단위)'].replace(age_code)
```

```python
df_sight = df.groupby(['연령대', '성별코드']).agg({'시력(좌)': 'mean', '시력(우)': 'mean'}).reset_index()
sns.barplot(data=df_sight, x="시력(좌)", y="연령대", hue="성별코드", errorbar=None)
```
