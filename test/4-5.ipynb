```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import koreanize_matplotlib

df=pd.read_csv("data/국가_온실가스_종류별_배출량_추이_20231031132400.csv", encoding="cp949")
df.head()
```

```python
df.rename(columns={"구분(1)":"종류"}, inplace=True)
df
```

```python
df_group=df.groupby("종류").mean()
df_group
```

```python
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
categories = ['CH₄', 'CO₂', 'HFCs', 'LULUCF', 'N₂O', 'PFCs', 'SF6']
data = {
    'CH₄': [27.537, 27.360, 27.323, 27.390, 27.840, 28.009, 27.249, 27.145],
    'CO₂': [636.638, 629.870, 634.321, 637.428, 650.220, 664.976, 643.767, 599.813],
    'HFCs': [8.095, 8.538, 7.931, 7.366, 9.649, 9.305, 6.855, 6.662],
    'LULUCF': [-45.016, -45.866, -46.561, -46.934, -41.689, -40.334, -37.710, -37.876],
    'N₂O': [12.999, 13.096, 13.024, 12.993, 13.833, 14.330, 14.244, 14.097],
    'PFCs': [2.321, 2.427, 1.521, 1.489, 2.123, 3.180, 3.003, 3.401],
    'SF6': [9.708, 10.798, 8.506, 7.022, 6.981, 7.178, 6.096, 5.106]
}

for category in categories:
    plt.plot(years, data[category], marker='o', label=category)

plt.xlabel('연도')
plt.ylabel('배출량')
plt.title('국내 온실가스 종류별 배출량 (2013-2020)')
plt.xticks(years)
plt.legend()
plt.grid(True)

```

```python
df = pd.DataFrame(data)

plt.figure(figsize=(15, 6))
sns.heatmap(df, annot=True, fmt=".2f", cmap="YlGnBu", xticklabels=years, yticklabels=categories)
plt.xlabel('연도')
plt.ylabel('종류')
plt.title('국내 온실가스 종류별 배출량 (2013-2020)')
plt.show()
```
