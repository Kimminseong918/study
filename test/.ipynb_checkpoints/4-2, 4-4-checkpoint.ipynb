```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import koreanize_matplotlib

df = pd.read_csv(
    "https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/online_oversea_sale_202210.csv.zip", 
    encoding="cp949")
```

```python
df["연도"] = df["시점"].apply(lambda x: x.split('.')[0])
df["연도"]=df["연도"].astype(int)

df["분기"] = df["시점"].apply(lambda x: x.split('.')[1][:1])
df["분기"] = df["분기"].astype(int)
```

```python
df_pivot=pd.pivot_table(df, index="국가(대륙)별", columns="연도", values="데이터", aggfunc='sum')
df_pivot
```

```python
df_pivot.T.plot(kind='bar', stacked=True)

plt.title('연도별 판매액')

plt.xlabel('국가(대륙)별')
plt.ylabel('데이터')

plt.legend(bbox_to_anchor=(1.5, 1), loc=2, borderaxespad=0.)
plt.legend(title="백분율")

plt.show()
```

```python
sns.relplot(data=df, x="연도", y="데이터", hue="국가(대륙)별", kind="line")
```

```python
contry=["미국", "중국", "일본"]
product=["의류 및 패션 관련상품", "화장품", "음반·비디오·악기"]
test4=df[(df['국가(대륙)별'].isin(contry) & df['상품군별'].isin(product))]

test4_p=test4.pivot_table(index=['국가(대륙)별', '상품군별', '연도'])

plt.figure(figsize=(15,6))
plt4= sns.FacetGrid(test4.reset_index(), col="국가(대륙)별", hue="상품군별", col_wrap=3, height=5, aspect=1.5)
plt4.map(sns.lineplot, "연도", "데이터")
plt4.add_legend(bbox_to_anchor=(1.03, 0.5), loc='center')
plt4.set_axis_labels("연도", "데이터")
plt4.set_titles(col_template="국가(대륙)별 = {col_name}")
plt.tight_layout()
plt.show()
```
