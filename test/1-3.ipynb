# 2.1 데이터 분석을 위한 파이썬 속성 코스
* True ==1, False==0 /// True !="1" (True는 문자 1과 다르다)
### strings
* lower: 문자열을 모두 소문자로
* upper: 문자열을 모두 대문자로
### lists
* .append(): 원소 추가
* list_name[ n ] : n번째 인덱스 출력
### Control Flow
* for i in range(3):
        print(lang[i])
  => 횟수를 지정하고 i에 해당하는 값 출력
* enumerate: 인덱스 번호와 원소를 동시에 가져온다
  for i, val in enumerate(lang):
      print(i, val) #value 값을 반드시 가져와야 한다
### 문자열
* strip() : 문자옆 앞 뒤 공백 제거, 다시 문자열에 넣어줘야 전처리 된 값 출력
* split() : 공백으로 문자열 분리
* 문자열[n:n] : 문자열 위치 지정하여 가져오기
* startswith : 문자열이 특정 문자를 포함하는지 확인
* in : 문자열에 특정 문자가 포함하는지 확인, T/F로 출력

# 2.2 판다스 치트시트를 활용한 기초 익히기
* import pandas as pd
### DataFrame
* 2차원의 자료 구조
* df=pd.DataFrame(  값 입력  )
### Series
* 1차원의 자료 구조
* df=[ " 컬럼 값 입력 " ] : 출력
* df[[ " 컬럼 값 입력 " ] : 표 형태로 출력
### Subset
* 두 개 이상의 컬럼을 가져 올 경우, 컬럼 값을 리스트 형태로 전환
### Summarize Data
* df["컬럼 명"].value_counts()
 : 빈도수 계산
### Reshaping
* df["컬럼명"].sort_values: 컬럼 기준으로 데이터 정렬
* df.sort_values("컬럼 명", ascending=False): 컬럼명 기준으로 역순으로 정렬
* df.drop(["컬럼 명"], axis= 1): 컬럼을 기준으로 drop하는 경우
### Group Data
* df.groupby((["컬럼 명 1"])["컬럼 명 2"].describe()
 => 컬럼 1을 컬럼 2와 그룹화 하여 값 출력
* pd.pivot_table(): 피벗 테이블 출력
### Ploting
* df.plot.원하는 그래프 모형()
