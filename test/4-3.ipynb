```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import koreanize_matplotlib

df = pd.read_csv(
    "https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/online_oversea_sale_202210.csv.zip", 
    encoding="cp949")
```

```python
df["연도"] = df["시점"].apply(lambda x: x.split('.')[0])
df["연도"]=df["연도"].astype(int)

df["분기"] = df["시점"].apply(lambda x: x.split('.')[1][:1])
df["분기"] = df["분기"].astype(int)
```

```python
df_test3=df[df['연도']==2020]

test3=pd.DataFrame(df_test3.groupby('국가(대륙)별')['데이터'].sum())

test3_unstack=test3.unstack()
```

```python
test3_unstack.plot(kind='bar', stacked=True, figsize=(15, 6))

plt.xlabel('지역')
plt.ylabel('판매액')
plt.legend(title='해외 판매 상품군', bbox_to_anchor=(1, 1))

plt.title('2020년의 온라인 해외판매 상품군 지역별 합계')
plt.show()
```
